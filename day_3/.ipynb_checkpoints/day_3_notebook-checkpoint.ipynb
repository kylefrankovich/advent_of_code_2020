```python
start_coords = [0,0]
len_tree_map = len(tree_map) - 1
positions = list(range(0,31))
count = 0

while len_tree_map > 0:
    # update start coords:
    start_coords[0] = start_coords[0] + 1
    start_coords[1] = start_coords[1] + 3
    
    # count if tree encounter
    if tree_map[start_coords[0]][positions[start_coords[1] % len(positions)]] == '#':
        count += 1
    
    # reduce number of steps:
    len_tree_map -= 1

print('number of tree encounters:', count) # 184 is correct!
```

### --- Part Two ---

Time to check the rest of the slopes - you need to minimize the probability of a sudden arboreal stop, after all.

Determine the number of trees you would encounter if, for each of the following slopes, you start at the top-left corner and traverse the map all the way to the bottom:

Right 1, down 1.
Right 3, down 1. (This is the slope you already checked.)
Right 5, down 1.
Right 7, down 1.
Right 1, down 2.
In the above example, these slopes would find 2, 7, 3, 4, and 2 tree(s) respectively; multiplied together, these produce the answer 336.

What do you get if you multiply together the number of trees encountered on each of the listed slopes?

```python
def slope_check(slope_val_1, slope_val_2, tree_map):
    start_coords = [0,0]
    len_tree_map = len(tree_map) - 1
    positions = list(range(0,31))
    count = 0

    while len_tree_map > 0:
        # update start coords:
        start_coords[0] = start_coords[0] + slope_val_2
        start_coords[1] = start_coords[1] + slope_val_1

        # count if tree encounter
        if tree_map[start_coords[0]][positions[start_coords[1] % len(positions)]] == '#':
            count += 1
#             print(count) # hacky way to get the solution when going down 2 jumps (index out of range)

        # reduce number of steps:
        len_tree_map -= 1

    print('number of tree encounters:', count) # 184 is correct!

slope_check(3,1,tree_map)
```

```python
slope_check(1,1,tree_map)
```

```python
slope_check(5,1,tree_map)
```

```python
slope_check(7,1,tree_map)
```

```python
slope_check(1,2,tree_map)
```
